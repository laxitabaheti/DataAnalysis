```python
k=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']
from scipy.stats import pearsonr


def function(x,y):
  z = pearsonr(x,y)
  return z


L=[]

for i in k:
  for j in k:
    if i==j:
      print(" ")
    else:
      x=df[i] 
      y=df[j]
      M=function(x,y)
      L.append(M)
      print(i,j)
      print(function(x,y))      
```

- pH and  citric acid ( - 0.5419041447395097 )
- alcohol and density ( - 0.49617977024170107 )

###Pearson Rank###
 
Rank varies between - 1 to + 1.

   *  +1 = strong positive relation between two variables
   *  -1 = strong negative relation between two variables
   * 0 =  no relation 

positive relation
- fixed acidity  and citric acid ( x=0.6717034347641061 )
- fixed acidity and density ( 0.6680472921189743 )
- free sulfur dioxide and total sulfur dioxide ( 0.6676664504810215 )
- alcohol and quality ( 0.47616632400113607 )

negative relation
- fixed acidity and pH ( - 0.6829781945685314 )
- volatile acidity and citric acid ( - 0.55249568455958 )
- pH and  citric acid ( - 0.5419041447395097 )
- alcohol and density ( - 0.49617977024170107 )
